```python
# 所以，綜觀以上兩個結果，我們認為，因為setAB_AC的長度 =1, 所以 A->B->C 相當可能。
# 以上是以「捨棄的角度來看」。
# 以下我們用「新增的角度來看」
```

```python
# 運算 setB-setA (命名為 setA=B_A), 表示大A(點數多)，到中B(點數中等)的變化過程中，新增的部分
setB_A = setB.difference(setA)
print(setB_A)
print("setB_A的長度 ={}".format(len(setB_A)))

print("--------------------------------")

# 運算 setC-setA (命名為 setC_A), 表示大A(點數多)，到小C(點數少)的變化過程中，新增加的部分
setC_A = setC.difference(setA)
print(setC_A)
print("setC_A的長度 ={}".format(len(setC_A)))
```
